```python
df_merged = df_merged.drop(['company_profile', 'requirements','benefits'], axis=1)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
```

```python
df_merged.fraudulent.value_counts()
```

```python
df_merged.columns
```

```python
x = tfidf_vectorizer.fit_transform(df_merged['description'])
y = df_merged['fraudulent']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=20)
```

```python
from sklearn.utils.class_weight import compute_class_weight
class_weights = compute_class_weight(class_weight = "balanced" , classes = np.unique(y_train), y = y_train)
class_weights_dict = dict(zip(np.unique(y_train), class_weights))
```

```python
svm_model = SVC(kernel='linear', C=4, random_state=20)
svm_model.fit(x_train, y_train)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score

# Define the model using hinge loss (equivalent to SVM)
svm_sgd = SGDClassifier(loss='hinge', learning_rate='optimal', random_state=20)

epochs = 1 # Define number of epochs
for epoch in range(epochs):
    svm_sgd.partial_fit(x_train, y_train, classes=[0, 1])  # Incremental learning
    train_accuracy = accuracy_score(y_train, svm_sgd.predict(x_train))
    print(f"Epoch {epoch + 1}/{epochs} - Training Accuracy: {train_accuracy:.4f}")

# Final evaluation on test set
test_accuracy = accuracy_score(y_test, svm_sgd.predict(x_test))
print(f"Final Test Accuracy: {test_accuracy:.4f}")
```

```python
y_pred = svm_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
print("\nClassification Report:\n", report)
print("\nConfusion Matrix:\n", confusion)
```

```python
def job_post_detection():

    new_job_post = input("Enter the job description: ")

    # Put the new_job_post string inside a list to make it an iterable
    new_job_post_vectorized = tfidf_vectorizer.transform([new_job_post])
    prediction = svm_model.predict(new_job_post_vectorized)

    if prediction[0] == 1:
        print("\nThis job posting is predicted to be FAKE.")
    else:
        print("\nThis job posting is predicted to be REAL.")
job_post_detection()
```
